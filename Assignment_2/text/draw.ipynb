```python
import matplotlib.pyplot as plt
```

```python
doc = [10, 20, 30, 40, 50]
precision = [0.70, 0.60, 0.53, 0.50, 0.44]

plt.figure(figsize=(8, 6))
plt.plot(doc, precision, marker='o', linestyle='-', color='b', label='Precision Curve')

plt.title('Precision Graph', fontsize=16)
plt.xlabel('Number of documents', fontsize=14)
plt.ylabel('Precision', fontsize=14)

plt.xlim([0, 60])
plt.ylim([0, 1.0])

plt.grid(True, linestyle='--', alpha=0.6)

plt.legend(loc='upper right')

plt.show()
```

```python
recall = [0.07, 0.12, 0.16, 0.20, 0.22]
precision = [0.70, 0.60, 0.53, 0.50, 0.44]

plt.figure(figsize=(8, 6))

plt.plot(recall, precision, marker='o', linestyle='-', color='b', label='Precision-Recall Curve')

plt.title('Precision-Recall Graph', fontsize=16)
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)

plt.xlim([0, 0.25])
plt.ylim([0, 1.0])

plt.grid(True, linestyle='--', alpha=0.6)

plt.legend(loc='upper right')

plt.show()
```
